### Mann-Kendal test

La prueba no paramtrica de Mann-Kendall (MK) es ampliamente utilizada en el anlisis de tendencias debido a su robustez y simplicidad. El estadstico de prueba $(S)$ de la serie $p_1,p_2,p_3,..., p_n$ se estiman a travs de la ecuacin: 

\begin{equation}
S=\sum_{j=1}^{n-1}\sum_{k=j+1}^n \ sign   (p_k-p_j),
\end{equation} 


 donde $n$ representa la observacin total, $p_k$ y $p_j $indica los datos observados del tiempo $k$ y $j$, donde la funcin $sign$ se define como:

\begin{equation}
\ sign  (p_k-p_j)= \left\{ \begin{array}{lcc}
 1 & \mbox{if} & (p_k-p_j)>0 \\ \\ 0 & \mbox{if} & (p_k-p_j)=0 \\ \\ -1 & \mbox{if} & (p_k-p_j)<0 \end{array} \right.
\end{equation}


La varianza de $S$ se estima de la siguiente manera:

\begin{equation}
VAR(S)=\frac{1}{18}\left[   n(n-1)(2n+5)-\sum_{j=1}^g T_j(T_j-1)(2T_j+5).  \right]
\end{equation}

Aqu, $g$ es el nmero del grupo vinculado, $T_j$ se refiere a la extensin del $j$-simo nmero vinculado. A partir de $S$ y *VAR(S)*, la medida de prueba estandarizada $Z$ se calcula utilizando la ecuacin:

\begin{equation}\label{Z}
Z = \left\{ \begin{array}{lcc}
\frac{S-1}{\sqrt{VAR(S)}} & \mbox{if} & S > 0 \\ 
\\ 0 & \mbox{if} & S=0 \\ 
\\ \frac{S+1}{\sqrt{VAR(S)}} & \mbox{if} & S<0 \end{array} \right.
\end{equation}

Un valor positivo de $Z$ indica tendencias crecientes, mientras que un valor negativo indica tendencias decrecientes en los datos de la serie temporal.



   * $H_0$: No hay una tendencia montona en la serie.
   
   * $H_a$: Existe una tendencia. Esta tendencia puede ser positiva, negativa o no nula.

Las pruebas de Mann-Kendall devuelven una tupla con nombre que contiene:

* Tendencia: indica la tendencia (creciente, decreciente o sin tendencia)

* h: Verdadero (si la tendencia est presente) o Falso (si la tendencia est ausente)
     
* p: valor p de la prueba de significancia
     
* z: estadsticas de prueba normalizadas
     
* Tau: Kendall Tau
     
* s: puntuacin de Mann-Kendal
     
* var_s: Varianza S
     
* pendiente: estimador de Theil-Sen/pendiente
     
* interseccin: interseccin de Kendall-Theil Robust Line, para pruebas estacionales, ciclo de perodo completo considerado como paso de tiempo unitario

El mtodo de ITA fue introducido por Sen Zekai en 2011 para evaluar las tendencias de diferentes datos de series temporales. Este mtodo no considera la normalidad, la autocorrelacin serial ni la longitud de los datos. 

El ITA divide la serie temporal en dos partes iguales, las ordena en orden ascendente y las representa grficamente en un sistema de coordenadas cartesianas. El eje $X$ del sistema de coordenadas cartesianas se utiliza para la primera mitad de la serie, mientras que el eje $Y$ se utiliza para la segunda mitad. La lnea 1:1 en el grfico de correlacin indica que no hay tendencia, la lnea de datos por encima de la lnea 1:1 indica tendencias crecientes/positivas y por debajo de la lnea indica tendencias negativas/decrecientes.
