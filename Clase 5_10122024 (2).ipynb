```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv")
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
X=data[['NO','PM25']]
y=data['PM10'] #variable objetivo
```

```python
split_index=int(len(data)*0.8)
X_train=X.iloc[:split_index]
X_test=X.iloc[split_index:]
y_train=y.iloc[:split_index]
y_test=y.iloc[split_index:]
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
svr_model=SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr_model.fit(X_train_scaled, y_train)
```

```python
y_test_pred=svr_model.predict(X_test_scaled)
```

```python
y_test_pred
```

```python
test_mse=mean_squared_error(y_test,y_test_pred)
test_r2=r2_score(y_test,y_test_pred)
```

```python
test_mse  #error cuadratico medio
```

```python
test_r2   #coeficiente de determinacion
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.iloc[split_index:]['fecha'].to_numpy(), y_test,label='valores reales',color='green')
plt.plot(data.iloc[split_index:]['fecha'].to_numpy(), y_test_pred,label='predicciones',color='red')
plt.xlabel('Fecha')
plt.ylabel('PM10')
plt.legend()
plt.xticks(data.iloc[split_index:]['fecha'][::15],rotation=45)
plt.tight_layout()
plt.show()
```

```python
fechas = data.iloc[split_index:]['fecha']  
y_test = y_test  
y_test_pred = y_test_pred  


print(f"Fechas: {len(fechas)}, Valores reales: {len(y_test)}, Predicciones: {len(y_test_pred)}")


import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(fechas, y_test, label='Valores Reales', linestyle='-', color='blue')
plt.plot(fechas, y_test_pred, label='Predicciones', linestyle='--', color='red')
plt.title('Valores Reales vs Predicciones')
plt.xlabel('Fecha')
plt.ylabel('CO')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test,y_test_pred,color='blue')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],"r--",label='linea ideal')
plt.title('valores predictivos vs reales')
plt.xlabel('valores reales')
plt.ylabel('valores predichos')
```
